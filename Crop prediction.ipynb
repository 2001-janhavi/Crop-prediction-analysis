```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Janhavi/Downloads/Crop Production data.csv")
```

```python
print(df.isnull().sum())
```

```python
mean_production = df['Production'].mean()
```

```python
#changing the null values with mean of production
df['Production'].fillna(mean_production, inplace=True)
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Area'], bins=20, kde=True)
plt.title('Distribution of Area')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Crop', order=df['Crop'].value_counts().index)
plt.title('Crop Distribution')
plt.xticks(rotation=90)
plt.xlabel('Crop')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Area', y='Production', alpha=0.5)
plt.title('Area vs Production')
plt.xlabel('Area')
plt.ylabel('Production')
plt.show()






```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='State_Name', order=df['State_Name'].value_counts().index)
plt.title('State Distribution')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Season', order=df['Season'].value_counts().index)
plt.title('Season Distribution')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = df[['Area', 'Production']].corr()

# Heatmap of correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
